–î–æ–ø—É—Å—Ç–∏–º, –º—ã –ø—Ä–æ–≤–æ–¥–∏–º –æ–Ω–ª–∞–π–Ω-–∫–æ–Ω–∫—É—Ä—Å —Ä–∞–±–æ—Ç –º–æ–ª–æ–¥—ã—Ö —Ö—É–¥–æ–∂–Ω–∏–∫–æ–≤. –í—Å–µ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–æ `N` —Ä–∞–±–æ—Ç, –∫–æ—Ç–æ—Ä—ã–µ –∏–¥–µ–Ω—Ç–∏—Ñ–∏—Ü–∏—Ä—É—é—Ç—Å—è —á–∏—Å–ª–∞–º–∏ –æ—Ç `0` –¥–æ `N-1` –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –ù—É–∂–Ω–æ –ø–æ–¥–¥–µ—Ä–∂–∞—Ç—å 3 —Ç–∏–ø–∞ –∑–∞–ø—Ä–æ—Å–∞:

*–î–ª—è —Å–∞–º–æ–≥–æ –ø—Ä–æ—Å—Ç–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ –∏ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏.*

```python
class Competition:
    def _change_score(self, participant_id, change):
        self._scores[participant_id] += change

    def __init__(self, participant_count):
        self._scores = [0] * participant_count

    def like(self, participant_id):
        self._change_score(participant_id, 1)

    def dislike(self, participant_id):
        self._change_score(participant_id, -1)

    def get_best_works(self, count):
        scores_ids = [(value, id) for id, value in enumerate(self._scores)]
        scores_ids.sort(reverse=True)
        return [id for _, id in scores_ids[:count]]
```

–ú–æ–∂–Ω–æ –ø—Ä–µ–¥–ª–æ–∂–∏—Ç—å –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ —Å–æ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–æ–π —Å—Ç—Ä—É–∫—Ç—É—Ä–æ–π –¥–∞–Ω–Ω—ã—Ö.

```python
from sortedcontainers import SortedSet

class Competition:
    def __init__(self, participant_count):
        self._scores = [0] * participant_count
        initial_frequencies = [(0, id) for id in range(participant_count)]
        self._ordered_works = SortedSet(initial_frequencies)

    def _change_score(self, participant_id, change):
        old_participant_stat = (self._scores[participant_id], participant_id)
        self._ordered_works.remove(old_participant_stat)
        
        self._scores[participant_id] += change
        
        new_participant_stat = (self._scores[participant_id], participant_id)
        self._ordered_works.add(new_participant_stat)

    def like(self, participant_id):
        self._change_score(participant_id, 1)

    def dislike(self, participant_id):
        self._change_score(participant_id, -1)

    def get_best_works(self, count):
        return [id for (_, id) in self._ordered_works[-count:][::-1]]
```

–ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ `get_best_works()` –ø–µ—Ä–≤–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –º–æ–∂–Ω–æ —Å–æ–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å, –∑–∞–º–µ–Ω–∏–≤ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫—É –Ω–∞ –∞–ª–≥–æ—Ä–∏—Ç–º `QuickSelect`, –∫–æ—Ç–æ—Ä—ã–π –¥–∞—Å—Ç –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã `O(N)`, –Ω–æ —ç—Ç–æ –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –Ω–µ —Ç–∞–∫ –≤–∞–∂–Ω–æ, –ø–æ—Å–∫–æ–ª—å–∫—É –¥–∞–∂–µ `O(N)` –±–æ–ª—å—à–µ, —á–µ–º `O(K)`.


–í—Ç–æ—Ä–æ–µ —Ä–µ—à–µ–Ω–∏–µ –Ω–∞–º–Ω–æ–≥–æ –±—ã—Å—Ç—Ä–µ–µ –Ω–∞—Ö–æ–¥–∏—Ç –ª—É—á—à–∏–µ —Ä–∞–±–æ—Ç—ã. –í–æ–∑–º–æ–∂–Ω–æ, –µ–≥–æ —Å—Ç–æ–∏—Ç –ø—Ä–µ–¥–ø–æ—á–µ—Å—Ç—å, –µ—Å–ª–∏ —É –Ω–∞—Å –±–æ–ª—å—à–µ –Ω–µ—Ç –Ω–∏–∫–∞–∫–∏—Ö –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö. –ù–æ –º–æ–∂–µ—Ç –æ–∫–∞–∑–∞—Ç—å—Å—è, —á—Ç–æ —ç—Ç–æ—Ç —Ç–æ–ø –Ω—É–∂–Ω–æ –ø–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞—Ç—å —Ä–∞–∑ –≤ —Å—É—Ç–∫–∏, —á—Ç–æ–±—ã –Ω–µ –±—ã–ª–æ –Ω–∞–∫—Ä—É—Ç–∫–∏ –≥–æ–ª–æ—Å–æ–≤. –ü—Ä–∏ —ç—Ç–æ–º –ø–æ—Ç–æ–∫ —Ä–µ–∞–∫—Ü–∏–π –Ω–∞ —Ä–∞–±–æ—Ç—ã –æ–∂–∏–¥–∞–µ—Ç—Å—è –±–æ–ª—å—à–æ–π. –¢–æ–≥–¥–∞ –ø–µ—Ä–≤–æ–µ —Ä–µ—à–µ–Ω–∏–µ –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª—å–Ω–µ–µ, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –±—ã—Å—Ç—Ä–æ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –æ—Å–Ω–æ–≤–Ω–æ–π –ø–æ—Ç–æ–∫ –∑–∞–ø—Ä–æ—Å–æ–≤, –ø—Ä–∏ —ç—Ç–æ–º –Ω–µ —Å–æ–∑–¥–∞–≤–∞—è –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏ –º–µ–¥–ª–µ–Ω–Ω—ã—Ö —á–∞—Å—Ç–µ–π.


–í —Ä–µ–∞–ª—å–Ω–æ–π —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–µ –∑–∞—á–∞—Å—Ç—É—é –Ω–µ –≤—Å–µ–≥–¥–∞ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ. –ü–æ—ç—Ç–æ–º—É —Ä–∞–∑—Ä–∞–±–æ—Ç—á–∏–∫—É –Ω—É–∂–Ω–æ –≤—ã—è—Å–Ω–∏—Ç—å —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è, –æ—Ü–µ–Ω–∏—Ç—å —Å–∏–ª—å–Ω—ã–µ –∏ —Å–ª–∞–±—ã–µ —Å—Ç–æ—Ä–æ–Ω—ã –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤ –∏ –≤—ã–±—Ä–∞—Ç—å –Ω–∞–∏–ª—É—á—à—É—é. –¢–æ –∂–µ —Å–∞–º–æ–µ –Ω–∞ —Å–æ–±–µ—Å–µ–¥–æ–≤–∞–Ω–∏–∏. –ò–Ω—Ç–µ—Ä–≤—å—é–µ—Ä —Ö–æ—á–µ—Ç —É—Å–ª—ã—à–∞—Ç—å –Ω–µ —Ç–æ–ª—å–∫–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ, –Ω–æ –∏ —Ç–æ, –∫–∞–∫ –≤—ã –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç–µ –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã.

*–í –∞–Ω–≥–ª–æ—è–∑—ã—á–Ω–æ–π –ª–∏—Ç–µ—Ä–∞—Ç—É—Ä–µ —á–∞—Å—Ç–æ —É–ø–æ—Ç—Ä–µ–±–ª—è–µ—Ç—Å—è —Ç–µ—Ä–º–∏–Ω `trade-off`, –∫–æ—Ç–æ—Ä—ã–π –º–æ–∂–Ω–æ –æ—Ç–¥–∞–ª—ë–Ω–Ω–æ –ø–µ—Ä–µ–≤–µ—Å—Ç–∏ –∫–∞–∫ `¬´–∫–æ–º–ø—Ä–æ–º–∏—Å—Å¬ª`. –û—Ç –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –Ω–∞ –∏–Ω—Ç–µ—Ä–≤—å—é –æ–∂–∏–¥–∞–µ—Ç—Å—è –Ω–µ –ø—Ä–æ—Å—Ç–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ, –Ω–æ –∏ —Ç–æ, –Ω–∞—Å–∫–æ–ª—å–∫–æ –æ–Ω –º–æ–∂–µ—Ç —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤—ã.*

# –ó–∞–¥–∞—á–∞ 2

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ –º–∞—Å—Å–∏–≤ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª. –û–Ω–∞ –Ω–∞—Ö–æ–¥–∏—Ç –ø—Ä–µ—Ñ–∏–∫—Å —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π —Å—É–º–º–æ–π –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –æ–¥–Ω–æ —á–∏—Å–ª–æ ‚Äî —Å—É–º–º—É —á–∏—Å–µ–ª —ç—Ç–æ–≥–æ –ø—Ä–µ—Ñ–∏–∫—Å–∞.


üí° –ü—Ä–µ—Ñ–∏–∫—Å–æ–º –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –Ω–∞–±–æ—Ä –ø–µ—Ä–≤—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏. –ü—Ä–∏–º–µ—Ä—ã:
–ú–∞—Å—Å–∏–≤ `[2, 0, 2, 1]` –∏–º–µ–µ—Ç –ø—Ä–µ—Ñ–∏–∫—Å—ã `[]` (–ø—É—Å—Ç–æ–π –ø—Ä–µ—Ñ–∏–∫—Å), `[2]`, `[2, 0]`, `[2, 0, 2]`, `[2, 0, 2, 1]`.
–°—Ç—Ä–æ–∫–∞ `cat` —Å–æ–¥–µ—Ä–∂–∏—Ç –ø—Ä–µ—Ñ–∏–∫—Å—ã `(–ø—É—Å—Ç–æ)`, `c`, `ca`, `cat`.

```python
def get_max_prefix_sum(arr):
    maximal_sum = 0
    for i in range(1, len(arr)):
        current_sum = sum(arr[0:i])
        maximal_sum = max(maximal_sum, current_sum)
    return maximal_sum
```
